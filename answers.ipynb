```
nyc_trip_distance = nyc_taxi_new['trip_distance']
#nyc_trip_distance.mean()
#nyc_trip_distance.min()
#nyc_trip_distance.max()
nyc_norm = (nyc_trip_distance - nyc_trip_distance.mean()) / (nyc_trip_distance.max() - nyc_trip_distance.min())
nyc_norm.head()
```

```
# calismadi
from sklearn import preprocessing

# Create x, where x the 'scores' column's values as floats
x = nyc_taxi_new[['trip_distance']].values.astype(float)

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(x)

# Run the normalizer on the dataframe
df_normalized = pd.DataFrame(x_scaled)

# View the dataframe
df_normalized

# Plot the dataframe
df_normalized.plot(kind='bar')
```
