```
# Question 3
# I checked the "trip_distance", "fare_amount", "passenger_count" and "total_amount" columns are zero or below. 
# Because these columns should not be zero or least, in my opinion
```

```
nyc_taxi_new = nyc_taxi.loc[ ~(( nyc_taxi['trip_distance'] <= 0 ) | ( nyc_taxi['fare_amount'] <= 0 ) | (nyc_taxi['passenger_count'] <= 0 ) | (nyc_taxi['total_amount'] <= 0 ))]
```

```
#Â Q4 - Normalizing the 'trip_distance'
```

```
# 04 - Before the normalization process we need to eliminate the outliers from dataset.
#Â For this elimination, we need to find the "mean" and "standard deviation-sd" values of data.
```

```
mean = nyc_taxi_new['trip_distance'].mean()
sd = nyc_taxi_new['trip_distance'].std()
print("mean:" , mean , ", sd:", sd)
```

```
# Q4 - Eliminating the outliers
```

```
nyc_taxi_last = nyc_taxi_new.loc[~ ((nyc_taxi_new['trip_distance'] > mean - 2 * sd) & (nyc_taxi_new['trip_distance'] > mean + 2 * sd)) ]
```

```
#Â Q4 - Normalizing
```

```
norm = (nyc_taxi_last['trip_distance'] - nyc_taxi_last['trip_distance'].mean()) / (nyc_taxi_last['trip_distance'].max() - nyc_taxi_last['trip_distance'].min())
```

```
# Q4 - Increasing the number of bins
```

```
bins = 100
# second option is when the "bins" values is 1000. When "bins" value increase the detail of graph is varied
# bins = 1000

graph = norm.plot(kind="hist", bins = bins)
# Customizing the "graph"
graph.set_xlim(-1,1)
graph.set_ylabel('Number of Rows')
graph.set_xlabel('Normalization')
graph.set_title('Graph of the normalizaiton of "trip_distance"')
```

```
# Q6 - Create new column which name is "payment_type_text"
```

```
nyc_taxi_last['payment_type_text'] = 0
```

```
# Q6 - Define the condition
#Â The condition turn "cash" if "payment_type" is 2, turns "other" for other values
```

```
condition = nyc_taxi_last['payment_type'] == 2
nyc_taxi_last.loc[condition, 'payment_type_text' ] = 'Cash'
nyc_taxi_last.loc[~condition, 'payment_type_text' ] = 'Other'
nyc_taxi_last.head()
```

```
# Question 7 - Correlation between leaving no tip and paying with cash
```

```
# Q7 - To find the correlation between "no tip" and "paying cash", I create new columns as "tip_correlation" and "cash_correlation".
# They represent 1 if "there is no tip" and "paid with cash". With this process if there is any correlation we can see.
# If we find the result close to 1, then we can say that there is a correlation between "no tip" and "paying cash" relation.
```

```
nyc_taxi_last['tip_correlation'] = 0
nyc_taxi_last['cash_correlation'] = 0
nyc_taxi_last.head()
```

```
condition_tip = nyc_taxi_last['tip_amount'] == 0
nyc_taxi_last.loc[condition_tip, 'tip_correlation' ] = 1

condition_cash = nyc_taxi_last['payment_type'] == 2
nyc_taxi_last.loc[condition_cash, 'cash_correlation' ] = 1
nyc_taxi_last.head()
```

```
nyc_taxi_last['tip_correlation'].corr(nyc_taxi_last['cash_correlation'])
```

```
# There is a correlation between "no tip" and "paying cash" relation.
```
